```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """Calculates the number of rice grains on each day and the cumulative number of rice grains.

    Parameters
    -----------
    n_days: int
        The number of days for which rice is received (default: 100)

    Returns
    -------
    list_n_grains : list
        A list of the number of rice grains received each day (length is equal to n_days)
    list_total_grains : list
        A list of cumulative rice grains received up to that day (length is equal to n_days)
    """
    list_n_grains = []
    list_total_grains = []
    total_rice = 0
    
    for day in range(1, n_days + 1):
        rice_today = 2 ** (day - 1)  # The number of rice grains on this day
        total_rice += rice_today
        list_n_grains.append(rice_today)
        list_total_grains.append(total_rice)
    
    # Plotting the graphs
    plt.figure(figsize=(10, 6))
    
    # Number of rice grains per day
    plt.subplot(2, 1, 1)
    plt.plot(range(1, n_days + 1), list_n_grains, label="Rice grains per day")
    plt.title("Rice Grains per Day")
    plt.xlabel("Day")
    plt.ylabel("Rice grains")
    plt.grid(True)

    # Cumulative rice grains
    plt.subplot(2, 1, 2)
    plt.plot(range(1, n_days + 1), list_total_grains, label="Cumulative rice grains", color='orange')
    plt.title("Cumulative Rice Grains")
    plt.xlabel("Day")
    plt.ylabel("Total rice grains")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return list_n_grains, list_total_grains

# Call the function for 100 days
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=100)

# Output the total rice on the 100th day
print(f"Total rice grains received on the 100th day: {list_n_grains[99]}")

"QUETION 2"
def compute_sorori_shinzaemon_dynamic(n_days=30):
    """Calculates the number of rice grains for any number of days.

    Parameters
    -----------
    n_days: int
        The number of days for which rice is received (default: 30)

    Returns
    -------
    list_n_grains : list
        A list of the number of rice grains received each day (length is equal to n_days)
    list_total_grains : list
        A list of cumulative rice grains received up to that day (length is equal to n_days)
    """
    list_n_grains = []
    list_total_grains = []
    total_rice = 0
    
    for day in range(1, n_days + 1):
        rice_today = 2 ** (day - 1)
        total_rice += rice_today
        list_n_grains.append(rice_today)
        list_total_grains.append(total_rice)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    
    # Number of rice grains per day
    plt.subplot(2, 1, 1)
    plt.plot(range(1, n_days + 1), list_n_grains, label="Rice grains per day")
    plt.title("Rice Grains per Day")
    plt.xlabel("Day")
    plt.ylabel("Rice grains")
    plt.grid(True)

    # Cumulative rice grains
    plt.subplot(2, 1, 2)
    plt.plot(range(1, n_days + 1), list_total_grains, label="Cumulative rice grains", color='orange')
    plt.title("Cumulative Rice Grains")
    plt.xlabel("Day")
    plt.ylabel("Total rice grains")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return list_n_grains, list_total_grains

# Call the function for 30 days
list_n_grains_30, list_total_grains_30 = compute_sorori_shinzaemon_dynamic(n_days=30)

# Output the total rice for the 30th day
print(f"Total rice grains received on the 30th day: {list_n_grains_30[29]}")


"QUETION 3"
def calculate_survival_days(total_rice, people_count):
    """Calculates how many days people can survive on the given amount of rice.

    Parameters
    -----------
    total_rice: int
        The total number of rice grains
    people_count: int
        The number of people

    Returns
    -------
    survival_days: int
        The number of days people can survive
    """
    rice_per_person_per_day = 100 / 0.02  # 100 grams of rice per person, 1 grain = 0.02 grams
    total_rice_needed_per_day = rice_per_person_per_day * people_count
    survival_days = total_rice // total_rice_needed_per_day
    return survival_days

# Example usage:
total_rice_received = list_total_grains[99]  # Rice received on the 100th day
people_count = 50
days_people_can_survive = calculate_survival_days(total_rice_received, people_count)
print(f"With {total_rice_received} grains of rice, {people_count} people can survive for {days_people_can_survive} days.")
```

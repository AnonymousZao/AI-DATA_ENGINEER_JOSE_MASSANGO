```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Generate random numbers
np.random.seed(0)
mean1 = [-3, 0]
covariance = [[1.0, 0.8], [0.8, 1.0]]
data1 = np.random.multivariate_normal(mean1, covariance, 500)

# Problem 2: Scatter plot visualization
plt.figure(figsize=(6, 6))
plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5, label="Dataset 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Generated Data")
plt.legend()
plt.show()

# Problem 3: Histogram visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(data1[:, 0], bins=30, alpha=0.7, color='b')
axes[0].set_title("Histogram of X-axis")
axes[1].hist(data1[:, 1], bins=30, alpha=0.7, color='r')
axes[1].set_title("Histogram of Y-axis")
plt.xlim(-6, 3)  # Set the same x-axis range for both
plt.show()

# Problem 4: Generate new dataset and scatter plot
mean2 = [0, -3]
data2 = np.random.multivariate_normal(mean2, covariance, 500)
plt.figure(figsize=(6, 6))
plt.scatter(data1[:, 0], data1[:, 1], alpha=0.5, label="Dataset 1", color='b')
plt.scatter(data2[:, 0], data2[:, 1], alpha=0.5, label="Dataset 2", color='r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Two Datasets")
plt.legend()
plt.show()

# Problem 5: Concatenate datasets
combined_data = np.vstack((data1, data2))
print("Shape of concatenated data:", combined_data.shape)

# Problem 6: Labeling
labels1 = np.zeros((500, 1))
labels2 = np.ones((500, 1))
labeled_data = np.hstack((combined_data, np.vstack((labels1, labels2))))
print("Shape of labeled data:", labeled_data.shape)
print("First 5 rows:\n", labeled_data[:5])
```

```python
import math
import matplotlib.pyplot as plt

def time_to_cover_volume(initial_volume, target_volume, doubling_time=5):
    """Calculates the time required to exceed the target volume with exponential growth."""
    time = 0
    volume = initial_volume
    times = []
    volumes = []
    
    while volume < target_volume:
        times.append(time)
        volumes.append(volume)
        volume *= 2
        time += doubling_time
    
    return time, times, volumes

def plot_growth(times, volumes, target_volume):
    """Plots the exponential growth of volume over time."""
    plt.figure(figsize=(10, 5))
    plt.plot(times, volumes, marker='o', linestyle='-')
    plt.axhline(y=target_volume, color='r', linestyle='--', label='Target Volume')
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume (cubic meters)")
    plt.title("Exponential Growth of Chestnut Manju Volume")
    plt.legend()
    plt.yscale("log")  # Log scale for better visualization
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

# Example parameters
chestnut_manju_volume = 0.0005  # Assumed volume of a chestnut manju in cubic meters
solar_system_volume = 1.2e39  # Estimated volume of the solar system in cubic meters

time_needed, times, volumes = time_to_cover_volume(chestnut_manju_volume, solar_system_volume)

print(f"Time required to fill the solar system: {time_needed} minutes")

plot_growth(times, volumes, solar_system_volume)
```
